# method-3, EXTRACTING EDGE FEATURES

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow
from skimage.filters import prewitt_h, prewitt_v
```

```python
# reading the image
img = imread("C:/Users/rijul/Downloads/istockphoto-523698315-1024x1024.jpg", as_gray= True)
```

```python
# calculating horizontal edges using prrewitt kernel
edges_prewitt_horizontal = prewitt_h(img)
```

```python
# calculating vertical edges using prewitt kernel
edges_prewitt_vertical = prewitt_v(img)
```

```python
imshow(edges_prewitt_vertical, cmap='gray')
```
